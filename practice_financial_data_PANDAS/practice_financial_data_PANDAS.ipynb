```python
# Dependencies
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
# Load a CSV file into a Pandas DataFrame

# Create Path to the CSV file
financial_data_path = Path("Resource/financial_statements.csv")

# Read the CSV into a df
financial_df = pd.read_csv(financial_data_path)

# print df
financial_df.head(10)
```

```python
# Perform basic exploration of the DataFrame, such as checking summary statistics and data types

# Use the 'describe()' method to provide summary statistics
sum_stats = financial_df.describe()

# Use the 'dtypes' attribute to provide data types
data_types = financial_df.dtypes
```

```python
# Print summary statistics
sum_stats
```

```python
# Print data types
data_types
```

```python
# Explore unique values in 'Company ' column (header has whitespace)
# print(financial_df.columns)

financial_df['Company '].value_counts()
```

```python
# Filter rows based on a specific condition in one or more columns

# Filter for 'Microsoft' and 'Market Cap' data worth (>=) a trillion dollars 
msft_market_cap_t_years = financial_df[(financial_df['Company '] == 'MSFT') & (financial_df['Market Cap(in B USD)'] >= 999.99)]
msft_market_cap_t_years
```

```python
# Group the data by a column and perform some aggregation functions (e.g., mean, sum) on other columns

# Group data by return on investment (ROI)
grouped_data = financial_df.groupby('ROI')

# Find average for the Inflation Rate(in US) on grouped_data
grouped_data['Inflation Rate(in US)'].mean()
```

```python
# Create a simple plot (line plot, bar chart, etc.) using the financial data.

# Create a line plot that visualizes earning per share for Apple over the last 13 years.
# Store earning per share data and corresponding years in variables
earning_per_share = financial_df[financial_df['Company ']=='AAPL']['Earning Per Share']
year = financial_df[financial_df['Company ']=='AAPL']['Year']

# Display line plot
plt.plot(year, earning_per_share, marker='o', linestyle='-')
plt.title("Apple - Earning Per Share (2009-2022)")
plt.xlabel("Year")
plt.ylabel("Earning Per Share")
plt.grid(True)
plt.show()
```
